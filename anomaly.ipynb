```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
grade_math = pd.read_csv('uk_math_scores_dataset.csv')
grade_math
```

```python
grade_math_anomaly = grade_math[(grade_math['Math_Score']<9) | (grade_math['Math_Score']>18)]
count = len(grade_math_anomaly) 
print(count)
```

```python
grade_math_anomaly
```

```python
plt.boxplot(grade_math['Math_Score'])
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=grade_math.index, y='Math_Score', data=grade_math, 
                hue=grade_math['Math_Score'].apply(lambda x: 'High/Low' if x > 18 or x < 9 else 'Normal'),
                palette={'High/Low': 'red', 'Normal': 'blue'}, 
                s=100, alpha=0.8)

plt.axhline(18, color='gray', linestyle='--', alpha=0.7)
plt.axhline(9, color='gray', linestyle='--', alpha=0.7)
plt.title('scatterplot of student grade', fontsize=16)
plt.xlabel('student number', fontsize=12)
plt.ylabel('score', fontsize=12)
plt.xticks([])
plt.legend(title='kind of score')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
